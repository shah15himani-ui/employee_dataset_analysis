Create Scatterplot of the "Age vs Employee Salary" feature

```python
plt.figure(figsize=(11,6))
sns.scatterplot(data = data, x = 'age', y = 'salary', color = 'blue', edgecolor = 'linen', alpha = 0.5)
plt.title("Scatterplot of Employee Age v/s Employee Salary")
plt.xlabel('Employee Age')
plt.ylabel('Employee Salary')
plt.show()
```

Create a Scatterplot of "Healthy Eating vs Active Lifestyle" feature

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data = data, x = 'healthy_eating', y = 'active_lifestyle', color = 'red',
                edgecolor = 'linen', alpha = 0.5)
plt.title("Scatterplot of Healthy Eating v/s Active Lifestyle")
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')
plt.show()
```

Create a Scatterplot of "Healthy Eating vs Salary"

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data = data, x = 'healthy_eating', y = 'salary', color = 'k',
                edgecolor = 'linen', alpha = 0.5)
plt.title("Scatterplot of Healthy Eating v/s Salary")
plt.xlabel('Healthy Eating')
plt.ylabel('Salary')
plt.show()
```

Create a Count Plot of "Groups" feature

```python
plt.figure(figsize=(10,5))
sns.countplot(data=data, x = 'groups', edgecolor = 'linen', alpha = 0.7)
plt.title("Count Plot of Groups")
plt.xlabel('Groups')
plt.ylabel('Count')
plt.show()
```

Create a Histogram of "Salary" feature

```python
plt.figure(figsize=(11,6))
sns.histplot(data=data, x = 'salary', color = 'orange', edgecolor = 'linen', alpha = 0.7, bins = 10)
plt.title("Histogram of Salary")
plt.xlabel('Salary')
plt.ylabel('Count')
plt.show()
```
